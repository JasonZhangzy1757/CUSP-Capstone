```python
from linearmodels.panel import PanelOLS
from linearmodels import RandomEffects
```

```python
year = pd.to_datetime(dfm['year'], format='%Y')
```

```python
year = pd.Categorical(year)
```

```python
print(year)
```

```python
dfm = dfm.set_index(['pop', 'year'])
```

```python
dfm.head()
```

```python
dfm['year'] = year
```

```python
dfm.head()
```

```python

from linearmodels.panel import PooledOLS
import statsmodels.api as sm
```

```python
exog_vars = ['households','income', 'year']
exog = sm.add_constant(dfm[exog_vars])
```

```python
mod = PooledOLS(dfm.gini, exog)
pooled_res = mod.fit()
print(pooled_res)
```

```python
from linearmodels.panel import RandomEffects
mod = RandomEffects(dfm.gini, exog)
re_res = mod.fit()
print(re_res)
```

```python
from linearmodels.panel import PanelOLS
exog_vars1 = ['households','income']
exog1 = sm.add_constant(dfm[exog_vars1])
```

```python
mod = PanelOLS(dfm.gini, exog1, entity_effects=True, time_effects=True)
fe_te_res = mod.fit()
print(fe_te_res)
```

```python
from linearmodels.panel import compare
print(compare({'RE':re_res,'Pooled':fe_te_res}))
```
