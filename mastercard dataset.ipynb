```python
industry=['eap', 'aut', 'ret', 'gro', 'app', 'acc']
```

```python
mastercardlist=[]
```

```python
for i in industry:
    temp=mastercard[mastercard['industry']==i]
    t1='txn_amt_'+i
    t2='txn_cnt_'+i
    t3='acct_cnt_'+i
    t4='avg_ticket_'+i
    t5='avg_freq_'+i
    t6='avg_spend_amt_'+i
    
    temp.columns=['cartodb_id', 'the_geom', 'fid', 'geoid', 'do_date', 'industry',
       'segment', 'geo_type', t1, t2, t3,t4,
       t5, t6]
    temp=temp.drop(columns=['cartodb_id', 'the_geom'])

    temp['geoid']=temp.geoid.apply(lambda x:x[0:11])
    temp=temp.groupby('geoid', as_index=False)[[t1, t2, t3,t4,t5, t6]].mean()
    mastercardlist.append(temp)
```
