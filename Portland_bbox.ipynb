```python
import geopandas as gpd
```

```python
gdf = gpd.read_file("CityIQ_Nodes_shp/CityIQ_Nodes.shp")
```

```python
gdf.head(20)
```

```python
x_min = 0
x_max = -180
y_min = 90
y_max = 0
```

```python
for i, row in gdf.iterrows():
    x_min = min(x_min, row['x'])
    x_max = max(x_max, row['x'])
    y_min = min(y_min, row['y'])
    y_max = max(y_max, row['y'])
```

```python
print(x_min, x_max, y_min, y_max)
```
